```python
from pii_inference import PIIDetector , mask_pii
checkpoint = "models/best_bert_bilstm_crf.pt"
detector   = PIIDetector(checkpoint)
```

```python

examples = [
    "my name is mariam amin what is your name and my phone number is 010-1234-5678",
    "Please send the documents to alice.jones@example.com by Friday.",
    "My SSN is 123-45-6789 and my credit card number is 4111-1111-1111-1111.",
    "You can reach me at +1-800-555-0199 or connect via MAC address 00:1A:2B:3C:4D:5E.",
    "Log in with username jsmith92 and password Hunter2!.",
    "The server IP is 192.168.1.105 and the IPv6 address is 2001:0db8:85a3::8a2e:0370:7334.",
    "The quick brown fox jumps over the lazy dog.",
    "my email is john.doe@example.com",
]

for text in examples:
    tagged = detector.predict(text)
    print(mask_pii(text, tagged))
```
